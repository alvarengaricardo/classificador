# Visualizando valores únicos

É importante visualizar os valores únicos das colunas quando trabalhamos com um grande volume de dados por algumas razões:

1) Detecção de erros e valores inconsistentes: Ao visualizar os valores únicos de uma coluna, podemos identificar rapidamente se há erros ou valores inconsistentes nos dados. Podemos encontrar valores estranhos, outliers ou categorias que não fazem sentido dentro daquela coluna específica. Essa visualização pode ajudar a identificar problemas de qualidade dos dados e tomar medidas corretivas.

2) Compreensão do domínio dos dados: Ao examinar os valores únicos de uma coluna, obtemos uma visão geral dos diferentes tipos ou categorias presentes nos dados. Isso nos ajuda a entender o domínio dos dados e a ter uma visão mais completa das informações contidas na coluna. Por exemplo, em uma coluna que representa os tipos de produtos vendidos, visualizar os valores únicos nos dá uma ideia dos diferentes tipos de produtos disponíveis.

3) Análise de frequência: Ao visualizar os valores únicos de uma coluna, também podemos analisar a frequência de cada valor. Isso nos permite ter uma noção de quais categorias são mais comuns e quais são mais raras. Essa informação pode ser útil para identificar padrões ou tendências nos dados e pode guiar análises subsequentes.

4) Pré-processamento de dados: A visualização dos valores únicos pode auxiliar no pré-processamento dos dados. Por exemplo, se uma coluna contém muitas categorias únicas e algumas delas têm uma frequência muito baixa, pode ser necessário agrupar essas categorias em uma única categoria denominada "Outros" para simplificar a análise. Essa visualização ajuda a identificar esses casos e facilita a decisão de como lidar com eles.

Em resumo, visualizar os valores únicos das colunas em um grande volume de dados é importante para identificar erros ou valores inconsistentes, compreender o domínio dos dados, analisar a frequência das categorias e facilitar o pré-processamento dos dados. Isso proporciona uma visão mais clara e ajuda a garantir a qualidade e a integridade dos dados utilizados na análise.

## Agrupando os dados

A função **groupby()** no pandas é usada para agrupar dados com base em uma ou mais colunas do DataFrame. Ela permite dividir o DataFrame em grupos com base nos valores únicos de uma ou mais colunas-chave e realizar operações agregadas em cada grupo.

A função **groupby()** segue os seguintes passos:

1) **Divisão:** O DataFrame é dividido em grupos com base nos valores exclusivos da(s) coluna(s) especificada(s) no argumento **by** da função **groupby()**. Cada grupo contém todas as linhas que possuem o mesmo valor na(s) coluna(s) de agrupamento.

2) **Aplicação:** Uma operação agregada é aplicada a cada grupo separadamente. Isso pode ser uma função de agregação embutida, como soma, média, contagem, mínimo ou máximo, ou uma função personalizada definida pelo usuário.

3) **Combinação:** Os resultados agregados de cada grupo são combinados e retornados como um novo objeto DataFrame.

O **groupby()** é uma poderosa ferramenta para análise de dados, permitindo segmentar e resumir os dados com base em diferentes critérios de agrupamento. Ele pode ser combinado com várias funções de agregação e operações de transformação para obter insights valiosos sobre os dados.

## Filtrando os registros

A filtragem de registros em um DataFrame é uma operação importante e comum na análise de dados. Ela permite selecionar um subconjunto de linhas do DataFrame com base em determinados critérios ou condições.

A filtragem de registros é útil quando você deseja extrair apenas os dados relevantes para uma determinada análise ou aplicar determinadas condições aos seus dados. Isso ajuda a reduzir o volume de dados e focalizar naqueles que atendem a determinadas condições.

A filtragem é geralmente realizada usando operadores de comparação, como igualdade (==), desigualdade (!=), maior que (>), menor que (<), maior ou igual a (>=), menor ou igual a (<=), ou operadores lógicos, como and, or e not. Esses operadores são usados em combinação com colunas do DataFrame e valores específicos para criar condições de filtro.

A filtragem de registros permite que você trabalhe com subconjuntos específicos de dados em seu DataFrame, facilitando a análise e manipulação de dados relevantes para suas necessidades específicas.

# Trabalhando com valores nulos

O tratamento de valores nulos ou ausentes em ciência de dados é de extrema importância por várias razões:

1) **Precisão dos resultados:** Valores nulos podem afetar negativamente a precisão e a confiabilidade dos resultados das análises. Quando os valores ausentes não são tratados, eles podem distorcer estatísticas descritivas, cálculos de média, desvio padrão, correlação, entre outros. O tratamento adequado desses valores ausentes ajuda a evitar resultados incorretos ou tendenciosos.

2) **Integridade dos dados:** Os valores nulos podem representar informações faltantes ou desconhecidas. Eles podem indicar problemas na coleta ou entrada de dados, erros de registro ou informações que não foram registradas. Tratá-los adequadamente permite que você mantenha a integridade dos dados, preenchendo os valores ausentes com dados relevantes ou identificando-os como informações desconhecidas.

3) **Compatibilidade com algoritmos e modelos:** Muitos algoritmos e modelos de aprendizado de máquina não lidam bem com valores nulos. A presença de valores ausentes pode levar a erros ou falhas durante a execução desses modelos. Portanto, é necessário tratar os valores nulos para garantir a compatibilidade e o bom desempenho dos algoritmos e modelos.

4) **Análise e visualização de dados:** Valores ausentes podem dificultar a análise e a visualização de dados. Por exemplo, ao criar gráficos ou visualizações, a presença de valores nulos pode resultar em pontos de dados ausentes ou em interpretações enganosas. O tratamento adequado de valores nulos permite que você aproveite ao máximo as ferramentas de análise e visualização de dados, garantindo uma representação precisa e compreensível dos seus dados.

Existem várias abordagens para tratar valores nulos, como preenchê-los com valores adequados, removê-los se forem insignificantes ou substituí-los por estatísticas resumidas, como média ou mediana. A escolha do método de tratamento depende do contexto do problema e da natureza dos dados. O objetivo é garantir que os valores nulos sejam tratados de forma apropriada e coerente com a análise ou modelo que está sendo aplicado.

```python
# verificando valores nulos na coluna DT_PSSE_BCO_INPO: .isnull()
len(funci[funci.DT_PSSE_BCO_INPO.isnull()])
# esses são os funcionários que tomaram posse no Banco do Brasil
```

```python
# filtrando os demais funcionários: notnull()
len(funci[funci.DT_PSSE_BCO_INPO.notnull()])
# esses são os funcionários provenientes de Bancos Incorporados
```

```python
# Algumas possíveis opções de tratamento de nulos
# Opção 1: excluir os registros nulos: .dropna()
# porém nesse caso estaríamos trabalhando com uma amostra muito pequena e pouco aleatória
len(funci.dropna(subset= ['DT_PSSE_BCO_INPO']))
```

```python
# Opção 2: excluir essa coluna com dados faltantes: .drop()
# dessa forma não perdemos nenhum registro, porém perderíamos uma informação que pode ser interessante para a nossa análise
len(funci.drop(columns=['DT_PSSE_BCO_INPO']))
```

```python
# Opção 3: transformar essa informação em uma coluna categórica chamada Indicador de Funci Incorporado (IN_FUN_INPO)
# 1 - funci oriundo de Bancos Incorporados
# 0 - funci não oriundo de Bancos Incorporados
# primeiro vamos preencher os valores nulos por 0: .fillna()
funci['IN_FUN_INPO'] = funci.DT_PSSE_BCO_INPO.fillna(0)
```

```python
# em seguida vamos substituir tudo que não for 0 para 1: .apply()
funci['IN_FUN_INPO'] = funci['IN_FUN_INPO'].apply(lambda x : 1 if not(x == 0) else x)
```

```python
# por fim, vamos excluir a coluna antiga: .drop()
funci.drop(columns=['DT_PSSE_BCO_INPO'], inplace=True)
```

```python
# checar se restam dados nulos no DataFrame: .isnull().any()
funci.isnull().any()
```

Cada prefixo (ou dependência) do Banco pode ser desmembrado em várias equipes e subequipes subordinadas umas às outras, as Unidades Organizacionais. 

Cada registro da tabela depe representa uma equipe, portanto existem vários registros de UOR para um mesmo prefixo.

Todos os funcionários estão vinculados a duas UOR, uma sendo a UOR principal do prefixo e outra a UOR de posição, representando a equipe à qual ele está vinculado. ([Humanograma](https://humanograma.intranet.bb.com.br/))

A UOR principal representa o prefixo como um todo. É a UOR de posição do cargo mais elevado dentro do prefixo (Gerente Geral, Diretor, etc.). Todas as outras UOR do prefixo estão vinculadas ou subordinadas a ela.
